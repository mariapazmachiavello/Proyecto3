**Region:** Uruguay/Mexico
**Categoria de Dominio:** Aspectos Sociales/Tasa de desempleo
**Pregunta a responder:** ¿Cómo ha variado la tasa de desempleo a lo largo de los años en Uruguay y México?

```python
import pandas as pd
import matplotlib.pyplot as plt

mexico = pd.read_csv("/content/Proyecto3/MEXICO_DESEMPLEO.csv",delimiter=",",encoding="latin1")
uruguay = pd.read_csv("/content/Proyecto3/URUGUAY_DESEMPLEO.csv",encoding="latin1")
```

```python
mexico.columns = ["Anio","Tasa"]
```

```python
mexico = mexico[mexico["Anio"].str.isnumeric()]
mexico["Anio"]=mexico["Anio"].astype(int)
mexico["Anio"].replace({"09/2023":"2023"})

#mexico["Tasa"]= mexico["Tasa"].str.replace('"','')
mexico["Tasa"]=mexico["Tasa"].astype(float)
```

```python
mexico_filtrado = mexico[(mexico["Anio"]<=2017)]
```

```python
plt.plot(uruguay["año"],uruguay["valor"],label="uruguay")
plt.plot(mexico_filtrado["Anio"],mexico_filtrado["Tasa"],label="mexico")

# Agregar leyenda al gráfico
plt.legend()

# Configuración adicional del gráfico
plt.title("Tasa de Desempleo en México y Uruguay")
plt.xlabel("Año")
plt.ylabel("Tasa de Desempleo")

# Mostrar el gráfico
plt.show()
```

Conclusiones:

```python
#La tasa de desempleo en Uruguay se destaca siendo más alta que en México durante la mayoría de los años en el período analizado.
#Para México, el año con la tasa de desempleo más alta fue 2009 (5.10%), mientras que en Uruguay, 2006 fue el año con la tasa más alta (10.8%).
#En México, la tasa de desempleo más baja se registró en 2019 (3.10%), mientras que en Uruguay, el año con la tasa más baja fue 2011 (6.0%).
```
